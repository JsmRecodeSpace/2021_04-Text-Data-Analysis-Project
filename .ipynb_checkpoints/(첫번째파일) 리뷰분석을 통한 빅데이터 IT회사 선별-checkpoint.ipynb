 - 조건을 만족하는 IT회사는 총 297개로 해당 IT회사들의 IT부서 리뷰를 크롤링

```python
total_enter_urls = []

for i in range(1, 100):
    url = f'https://www.jobplanet.co.kr/companies?industry_id=700&page={i}'
    res = requests.get(url)
    root = lxml.html.fromstring(res.text)
    enter_infos = root.cssselect('.content_col2_3.cominfo')
    enter_urls = []
    for j in range(len(enter_infos)):
        if int(enter_infos[j].cssselect('.row_end a:nth-of-type(1)')[0].text.split()[0]) >= 100:
            company_review_url = enter_infos[j].cssselect('.row_end a')[0].get('href')
            company_review_url2 = 'https://www.jobplanet.co.kr/' + company_review_url
            enter_urls.append(company_review_url2.split('%')[0] + '?occupation_1%5B%5D=11600')
    total_enter_urls.extend(enter_urls)

print('IT/웹/통신 기업중에서 총 리뷰수가 100개 이상인 기업의 수: ', len(total_enter_urls))
```

 - 셀레늄으로 리뷰 데이터수집

```python
dr = Chrome()
dr.get('https://www.jobplanet.co.kr/users/sign_in?_nav=gb')

# 로그인
email = dr.find_element_by_css_selector('#user_email')
email.send_keys('pnvr31p@dongguk.edu')
passward = dr.find_element_by_css_selector('#user_password')
passward.send_keys('cnldjq1!')
login_button = dr.find_element_by_css_selector('.btn_sign_up')
login_button.click()

time.sleep(2)
```

 - 크게 두 가지로 나누어 리뷰 정보를 스크랩.
     - 기업별 `통합 리뷰 정보`에서는 기업명, 기업평균점수를,
     - 기업별 `개인 리뷰 정보`에서는 근무부서, 현재근무여부, 근무지역, 리뷰 작성년도, 기업별점점수,  
       기업에 대한 한줄평, 기업 장점, 기업 단점, 기업에게 바라는 점, 기업 추천여부를 스크랩.

```python
# 기업별 통합 리뷰 정보
company_names = []; star_scores = [];
# 기업별 개인 리뷰 정보
departments = []; employees = []; locations = []; years =[]; scores = [];
evaluate_views = []; advantages = []; disadvantages = []; wishes = []; recommends = [];

# ----------- 기업별 통합 리뷰 정보 수집
for enter_url in total_enter_urls[:]:
    dr.get(enter_url)
    time.sleep(5)

    # 기업 이름
    company_name = dr.find_element_by_css_selector('h1.name a').text
    # 기업의 별점 점수
    star_score = float(dr.find_element_by_css_selector('span.rate_point').text)

    # 기업의 총 리뷰 개수
    total_reviews = int(dr.find_element_by_css_selector('#viewReviewsTitle.result span:nth-of-type(2)').text.replace(",",""))
    for i in range(total_reviews):
        company_names.append(company_name)
        star_scores.append(star_score)
        
    # ----------- 기업별 개인 리뷰 정보 수집
    # 기업별 모든 리뷰 모으면 스탑 -> 반복종료에 활용할 변수(num_review)
    num_review = 0
    while num_review != total_reviews:
        # 기업에 대한 품명 페이지당 5개
        ppl_reviews = dr.find_elements_by_css_selector('div.content_body_ty1')
        ppl_scores = dr.find_elements_by_css_selector('dd.sta_box')
        ppl_departs = dr.find_elements_by_css_selector('div.content_top_ty2')
        num_review += len(ppl_reviews)
        time.sleep(1)

        # 부서 / 현재직원여부 / 지역 / 리뷰 작성년도 / 기업평점
        # 한줄평 / 장점 / 단점 / 경영진에게 바라는 점 / 추천여부
        for i in range(len(ppl_reviews)):
            depart = ppl_departs[i].find_element_by_css_selector('span.txt1').text
            departments.append(depart)
            
            employee = ppl_departs[i].find_element_by_css_selector('span.txt1:nth-of-type(4)').text
            employees.append(employee)
            
            location = ppl_departs[i].find_element_by_css_selector('span.txt1:nth-of-type(6)').text
            if location.startswith('2'):
                locations.append('서울')
            else:
                locations.append(location)
            
            try:
                year = int(ppl_departs[i].find_element_by_css_selector('span.txt1:nth-of-type(8)').text.split('. ')[0])
                years.append(year)
            except:
                if location.startswith('2'):
                    year = int(location.split('. ')[0])
                    years.append(year)
                else:
                    years.append(2021)

            score = int(ppl_scores[i].find_element_by_css_selector('div.star_score').get_attribute('style').split(' ')[1].split('%')[0])
            scores.append(score)

            evaluate_view = ppl_reviews[i].find_element_by_css_selector('.us_label').text.strip('"')
            evaluate_views.append(evaluate_view)

            advantage = ppl_reviews[i].find_element_by_css_selector('.df1 span').text
            advantages.append(advantage)
 
            disadvantage = ppl_reviews[i].find_element_by_css_selector('.df1:nth-of-type(2) span').text
            disadvantages.append(disadvantage)

            wish = ppl_reviews[i].find_element_by_css_selector('.df1:nth-of-type(3) span').text
            wishes.append(wish)
            
            try:
                recommend = ppl_reviews[i].find_element_by_css_selector('p.txt.recommend.etc_box em').text
                recommends.append(recommend)
            except:
                recommends.append('비추천')
            time.sleep(0.5)
    

        # 다음페이지로 넘어가기
        dr.find_element_by_css_selector('.btn_pgnext').click()
        time.sleep(7)
        

    print(f'{company_name}회사, {total_reviews}개의 리뷰 수집 완료!')

    
```

```python
# --------------- 전처리
for i in range(len(evaluate_views)):
    evaluate_views[i] = evaluate_views[i].replace(',','.').replace('\n','.').replace('-','')
    advantages[i] = advantages[i].replace(',','.').replace('\n','.').replace('-','')
    disadvantages[i] = disadvantages[i].replace(",",".").replace("\n",".").replace('-','')
    wishes[i] = wishes[i].replace(",",".").replace("\n",".").replace('-','')
```

```python
# --------------- csv 파일로 저장
f = open('jobplanet_IT_department_views.csv', 'w', -1, 'utf-8')
f.write('company_name,star_score,department,employee,location,year,score,company_review,advantage,disadvantage,wish,recommend\n')
for i in range(len(company_names)):
    f.write(f'{company_names[i]},{star_scores[i]},{departments[i]},{employees[i]},{locations[i]},{years[i]},{scores[i]},{evaluate_views[i]},{advantages[i]},{disadvantages[i]},{wishes[i]},{recommends[i]}\n')
f.close()
```

# PART3. 데이터 전처리

 - 1. 다음 조건에 맞추어 개인 리뷰별 별점 점수를 이용하여 시각화와 감성분석에 활용할 라벨 변수(sentiment_label)를 생성.
     - 조건1: score가 80이상일 경우 기업에 대한 긍정의 반응으로 1 으로 설정
     - 조건2: score가 40이하일 경우 기업에 대한 부정의 반응으로 0 으로 설정
     - 조건3: score가 60일 때는, recommend가 추천일 경우 1로, 비추천일 경우 0 으로 설정

```python
# ---- 함수를 만들어 sentiment_label 변수 만들기

def makeLabel(data):
    data['sentiment_label'] = np.zeros(len(data))
    
    for i in range(len(data)):
        if data['score'][i] >= 80:
            data['sentiment_label'][i] = 1
            
        elif data['score'][i] <= 40:
            data['sentiment_label'][i] = 0
            
        else:
            if data['recommend'][i] == "추천":
                data['sentiment_label'][i] = 1
            else:
                data['sentiment_label'][i] = 0
    return data
```

```python
# 함수 적용과 int화
df = makeLabel(df)
df['sentiment_label'] = df['sentiment_label'].astype(int) 
```

```python
# 시각화에서 필요에 따라 타입변환 적용
df['score'] = df['score'].astype('category')
df['score'] = df['score'].cat.reorder_categories([20, 40, 60, 80, 100])
```

 - 2. 불필요한 칼럼들[star_score, department] 제거.

```python
# ---- 불필요한 칼럼 제거
df = df.drop(columns=['star_score', 'department', 'year'], axis=1)
```

 - 3. 분석하며 에러나는 NaN 값 처리

```python
df['company_review'] = df['company_review'].fillna('.')
```

```python
fig, axes = plt.subplots(1, 3, figsize=(10,5))
df['recommend'].value_counts().plot(kind='pie',autopct='%.2f%%', explode=(0,0.03), legend=True, colors=['lightcoral', 'lightblue'], textprops={'fontsize': 13}, ax=axes[0])
df['sentiment_label'].value_counts().plot(kind='pie',autopct='%.2f%%', explode=(0,0.03), labels=['부정','긍정'], legend=True, colors=['lightcoral', 'lightblue'], textprops={'fontsize': 13}, ax=axes[1])
sns.barplot(data=df.employee.value_counts().reset_index().rename(columns={'index':'workornot'}), y='employee', x='workornot', ax=axes[2])
axes[0].set_title('회사에 대한 추천/비추천 비율')
axes[1].set_title('회사에 대한 긍정/부정 비율')
axes[2].set_title('리뷰 남긴 사람들중 현재 근무여부')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
sns.countplot(data=df, x='score', hue='employee', palette='Blues', ax=axes[0])
sns.countplot(data=df, x='recommend', hue='employee', palette='Blues', ax=axes[1])
axes[0].set_title('직원별 기업점수')
axes[1].set_title('직원별 기업추천여부')
plt.show()
```

    4-2. 
 -   기업에 대한 개인별 리뷰 점수를 시각화하면
     왼쪽과 같다.  현재 근무를 하고 있는 직원들은
     기업 점수를 잘 주었고, 기업에 대해서도 추천을
     하고 있다. 

 -   추측하였던 대로 퇴사이후 새로운 일자리를 찾는
      사람들이 이전에 근무했던 곳에 대해서 낮은 점수를
      주고 비추천을 하는 것을 확인할 수 있었다.

 -   점수 시각화로 회사에 대한 긍부정에 있어서
     전직원들보다 회사에 남아있는 현직원일수록
     긍정의 비율이 더 높음을 알 수 있었다. 

```python
company_300th = df['company_name'].value_counts()[:300].index.tolist()
company_300th = df.query('company_name in @company_300th')

# axes[1]
crosstab_100th_3= pd.crosstab(company_300th.company_name, company_300th.sentiment_label, margins=True).sort_values('All', ascending=False)[1:].reset_index()
crosstab_100th_4= pd.crosstab(company_300th.company_name, company_300th.sentiment_label, margins=True, normalize='index')[:-1].reset_index()
crosstab_100th_5 = pd.merge(crosstab_100th_3, crosstab_100th_4, on='company_name').set_index('company_name').loc[:,['0_y','1_y']]
crosstab_100th_5 = crosstab_100th_5.rename(columns={'0_y':'부정', '1_y':'긍정'})

fig, axes = plt.subplots(2, 1, figsize=(20, 30))
crosstab_100th_5[:100].plot.bar(stacked=True, colormap='Greens', ax=axes[0])
crosstab_100th_5[100:300].plot.bar(stacked=True, colormap='Greens', ax=axes[1])
axes[0].axhline(y=0.5, color='r', linewidth=4)
axes[1].axhline(y=0.5, color='r', linewidth=4)
axes[0].set_title('기업 규모순으로 100개의 기업에 대한 긍부정 비율')
axes[1].set_title('기업 규모순으로 100개 이후 나머지 기업에 대한 긍부정 비율')
plt.show()
```

- 앞 부분 대기업들의 긍정 비율이 높긴 하지만 다른 회사들도 모두 살펴보았을 때 대기업이 아니어도 긍정의 비율이 높은 회사들이 많았다.  
  IT회사 규모가 클수록 IT부서의 긍정의 비율이 높을 가능성이 있지만 회사의 규모가 IT부서에 있어서 지배적인 요인은 아님을 알 수 있었다.

# PART5. WordCloud

 - IT회사의 `IT부서에 대한 장단점`은 무엇이 있는지 WordCloud로 살펴본다.
 - 실제 잡플래닛에서는 유료로 WordCloud를 제공중이다.  
   WordCloud에 나오는 단어를 클릭하면 그와 관련된 리뷰들이 나오도록 해놓았기에 그러한 점도 추가로 구현해 보기로 하였다.
***

-  명사 추출 함수 만들어 적용
     - wiki와 stanza에서 [명사만 추출 vs 동사만 추출 vs 명사와동사 추출] 비교
     - 둘 다 명사 추출에 있어서 비슷하였지만, wiki가 더 빠르고 단어 선택이 맘에 들어서 최종적으로 wiki 명사만 추출로 선택

```python
kiwi = Kiwi()
kiwi.prepare()

# 명사 추출하는 함수
def extract_noun(text):
    nouns = []
    result = kiwi.analyze(text)
    for token, pos, _, _ in result[0][0]:
        if pos[0] in 'N':
            yield token
```

-  장점 / 단점 TDM 만들기
     - 장단점 중 의미가 없어보이는 두 글자 단어들은 CountVectorizer의 stop_words인자로 제외시켰다.
     - '간', '함', '히'와 같이 한 글자인 단어들은 함수(drop_one_word)로 제외시켰다.

```python
cv = CountVectorizer(max_features=150, tokenizer=extract_noun,
                     stop_words=['가지', '다니', '나오', '다르', '대하','위하', '가능', '등등'])
# Stop Words 는 문서에서 단어장을 생성할 때 무시할 수 있는 단어를 말한다.
 
ad_tdm = cv.fit_transform(df['advantage'])
ad_word_count = pd.DataFrame({'단어': cv.get_feature_names(), '빈도': ad_tdm.sum(axis=0).flat})

disad_tdm = cv.fit_transform(df['disadvantage'])
disad_word_count = pd.DataFrame({'단어': cv.get_feature_names(), '빈도': disad_tdm.sum(axis=0).flat})
```

```python
def drop_one_word(word_count):
    drop_one_word_lst = []
    for word in word_count['단어'].unique().tolist():
        if len(word) != 1:
            drop_one_word_lst.append(word)

    word_count = word_count.query('단어 in @drop_one_word_lst')
    return word_count

ad_word_count = drop_one_word(ad_word_count)
disad_word_count = drop_one_word(disad_word_count)
```

 - 단어 빈도 카운트하여 단어구름 만들기


```python
ad_count_dic = ad_word_count.set_index('단어')['빈도'].to_dict()
disad_count_dic = disad_word_count.set_index('단어')['빈도'].to_dict()

# 회사원 이미지를 불러와 단어구름에 적용하여 본다.
mask = np.array(Image.open('person.png')) 
# 단어구름 만들기
ad_wc = WordCloud(font_path='C:/Users/82105/Downloads/NanumFontSetup_TTF_GOTHIC/NanumGothic.ttf', background_color='white',
               max_words=100, width=400, height=300, colormap = 'Blues', mask=mask)
disad_wc = WordCloud(font_path='C:/Users/82105/Downloads/NanumFontSetup_TTF_GOTHIC/NanumGothic.ttf', background_color='white',
               max_words=100, width=400, height=300, colormap = 'Reds', mask=mask)

ad_wc.generate_from_frequencies(ad_count_dic)
disad_wc.generate_from_frequencies(disad_count_dic)
plt.show()
```

#### 장점구름(part5)을 통해서  기업선별

 - 장단점별 단어구름을 만든 결과 장점 단점별로
      긍정적이거나 부정적인 단어들이 잘 분류되었다.
      하지만 그 중에는 ‘회사’, ‘야근’, ‘분위기’ 등과 같이  
      겹치는 단어들도 있음을 볼 수 있다. 
      같은 단어를 두고 다르게 말할 수 있음을 확인할
      수 있었고 구현한 함수로 검색하여 어떤 점에서
     장점이 될 수 있고 단점이 될 수  있는지 확인하였다. 
     
 - 장점 단어구름에 나온 단어들을 이용하여
      기업별로 장점 칼럼에서 몇개씩 나왔는지를 세고, 
      장점단어의 수가 전체 평균보다   
      높은 기업들 선별.
      변수 part5_WordCloud_positive_companies에 저장

-    단어구름 결과 선별된 기업의 수: 141

```python
# WordCloud로 나타낸 장점단어들의 파일을 불러온다
ad_word_count = pd.read_csv('id_department_advantage_count.csv', encoding='cp949')
ad_word_count = ad_word_count.sort_values('빈도', ascending=False)
ad_words = ad_word_count['단어'].tolist()

# ad_word_count라는 새로운 변수를 만들어 리뷰별 장점에 있는 단어의 수가 몇개 나왔는지 세어본다
df['ad_word_count'] = np.zeros(df.shape[0]).astype(int)
for word in ad_words:
    for i in range(len(df)):
        if word in df['advantage'][i]:
            df['ad_word_count'][i] += 1
# 기업 전체 리뷰에 장점에 있는 단어의 수가 평균적으로 몇개 나왔는지를 company_adWord_count_mean에 저장
company_adWord_count_mean = df.groupby('company_name')['ad_word_count'].mean().mean()

# company_adWord_count_mean가 전체 평균보다 높은 기업들 추출
part5_WordCloud_positive_companies = df.groupby('company_name').mean().sort_values('ad_word_count', ascending=False) \
    .query('ad_word_count > @company_adWord_count_mean').index.tolist()

print('단어구름 결과 선별된 기업수: ', len(part5_WordCloud_positive_companies))
```
